```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Load dataset
file_path = 'datasiap.xlsx'
data = pd.read_excel(file_path)
```

```python
# buat variable target
# 0: bukan keduanya, 1: Preeklampsia, 2: Eklampsia
data['target'] = 0
data.loc[data['_PE'] == 1, 'target'] = 1
data.loc[data['_PEB'] == 1, 'target'] = 2
```

```python
# hapus kolom target dan kolom yang tidak diperlukan
X = data.drop(['_PE', '_PEB', 'target'], axis=1)
y = data['target']
```

```python
# Split data menjadi train dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Standardkan features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# KNN model dengan k=5
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
# buat prediksi
y_pred = knn.predict(X_test)
```

```python
# evaluasi model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

```python
# tampilkan hasil
print(f'Accuracy: {accuracy:.2f}')
print('Classification Report:')
print(report)
```
